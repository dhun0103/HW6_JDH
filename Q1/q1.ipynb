```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def do():
    f1 = open('nationwide.csv')
    data1 = csv.reader(f1)
    next(data1)
    
    f2 = open('seoul.csv')
    data2 = csv.reader(f2)
    next(data2)
    
    f3 = open('dajeon.csv')
    data3 = csv.reader(f3)
    next(data3)
    
    f4 = open('busan.csv')
    data4 = csv.reader(f4)
    next(data4)
    
    f5 = open('jeju.csv')
    data5 = csv.reader(f5)
    next(data5)
    
    result1 = []
    result2 = []
    result3 = []
    result4 = []
    result5 = []
    
    seoul = []
    dajeon = []
    busan = []
    jeju = []
    cnt1 = 0
    cnt2 = 0
    cnt3 = 0
    cnt4 = 0
    
    for row in data1:
        if row[2] != '':
            result1.append(float(row[2]))
    
    for row in data2:
        if row[2] != '':
            result2.append(float(row[2]))
            seoul.append(float(row[2]) - result1[cnt1])
            cnt1+=1
            
    for row in data3:
        if row[2] != '':
            result3.append(float(row[2]))
            dajeon.append(float(row[2]) - result1[cnt2])
            cnt2+=1
            
    for row in data4:
        if row[2] != '':
            result4.append(float(row[2]))
            busan.append(float(row[2]) - result1[cnt3])
            cnt3+=1
            
    for row in data5:
        if row[2] != '':
            result5.append(float(row[2]))
            jeju.append(float(row[2]) - result1[cnt4])
            cnt4+=1
    
    
    
    plt.title('Average monthly temperatures in 2022')
    plt.xlabel("month")
    plt.ylabel("celsius")
    plt.xticks(np.arange(0,12,1), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dev'])
    plt.yticks([0, 5, 10, 15, 20, 25])
    
    plt.plot(result1, color='black', label='nationwide')
    plt.plot(result2, color='green', label='seoul')
    plt.plot(result3, color='red', label='dajeon')
    plt.plot(result4, color='pink', label='busan')
    plt.plot(result5, color='blue', label='jeju')
    plt.legend()
    plt.show()

    print("seoul")
    for i in range(12):
        print("%.1f ," % seoul[i], end='')
    print("\ndajeon")
    for i in range(12):
        print("%.1f ," % dajeon[i], end='')
    print("\nbusan")
    for i in range(12):
        print("%.1f ," % busan[i], end='')
    print("\njeju")
    for i in range(12):
        print("%.1f ," % jeju[i], end='')

    f1.close()
    f2.close()
    f3.close()
    f4.close()
    f5.close()
    
if __name__ == '__main__':
    do()
```
