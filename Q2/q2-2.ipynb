```python
def do(file, num):
    import csv
    import matplotlib.pyplot as plt
    
    f = open(file)
    data = csv.reader(f)
    
    result = []
    
    for row in data:
        if num == 100 and int(row[0]) == 100:
            for i in range(1,101):
                result.append(int(row[i]))
        if num == 1000 and int(row[0]) == 1000:
            for i in range(1,1001):
                result.append(int(row[i]))
        if num == 10000 and int(row[0]) == 10000:
            for i in range(1,10001):
                result.append(int(row[i]))
        if num == 100000 and int(row[0]) == 100000:
            for i in range(1,100001):
                result.append(int(row[i]))
    
    result.sort()
    
    xdata=[1,2,3,4,5,6,7]
    
    if num==100:
        plt.title('100 Dice trials')
    elif num==1000:
        plt.title('1000 Dice trials')
    elif num==10000:
        plt.title('10000 Dice trials')
    else:
        plt.title('100000 Dice trials')
    
    plt.xlabel('number')
    plt.ylabel('count')
    
    hist = plt.hist(result, xdata, color='green', edgecolor='black')
    
    plt.show()

    f.close()

if __name__ == '__main__':
    do("q2.csv", 100)
    do("q2.csv", 1000)
    do("q2.csv", 10000)
    do("q2.csv", 100000)
```
