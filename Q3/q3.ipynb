```python
import csv
import matplotlib.pyplot as plt

def do(file, year):
    f = open(file)
    data = csv.reader(f)
    next(data)
    
    male = 0
    female = 0
    
    for row in data:
        if '제주특별자치도  (5000000000)' in row[0]:
            male = int(row[3].replace(',',''))
            female = int(row[4].replace(',',''))
   

    if year==2008:
        plt.title('Gender of jeju in 2008')
    elif year==2012:
        plt.title('Gender of jeju in 2012')
    elif year==2016:
        plt.title('Gender of jeju in 2016')
    elif year==2020:
        plt.title('Gender of jeju in 2020')
    else:
        plt.title('Gender of jeju in 2022')
    
    plt.pie([male, female], labels=['male', 'female'], colors=['red', 'blue'], autopct='%.3f%%')
    
    plt.legend()
    plt.show()


    f.close()
    
if __name__ == '__main__':
    do('jeju2008.csv', 2008)
    do('jeju2012.csv', 2012)
    do('jeju2016.csv', 2016)
    do('jeju2020.csv', 2020)
    do('jeju2022.csv', 2022)
```
